1.Import only the necessary methods from the math module to calculate the sine of 90ยบ.
1.1. Import a very long named module like matplotlib.pyplot into a shorter version like plt so you can call its methods this way: plt.method().

```python
#Import
from math import sin

angle = 90

result = sin(angle)

# Print the result
print(result)

#1.1
# Import
import matplotlib.pyplot as plt

x = [1.1, 1.4, 1.5]
y = [1.2, 1.4, 1.6]
# Call its methods
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('plt Method')

plt.show()
```

2. Using the random module, calculate a random float between 0 and 2
2.1 What are the possible solutions of using randint(-4, 1)? And of randrange(-4, 1)?

```python
#2. Import
import random
random_float = random.uniform(0, 2)

print(random_float)
#Randint
random_ran = random.randint(-4,1)
print(random_ran)
#Randrange
random_range = random.randrange(-4,1)
print(random_range)
```
