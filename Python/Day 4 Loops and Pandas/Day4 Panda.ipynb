CSV file and XLSX file

```python
import pandas as pd

xlsx_path = 'movies_top_awards.xlsx'
df = pd.read_excel(xlsx_path)
# csv_path = 'movies_top_awards.csv'
# df = pd.read_csv(csv_path)
df.head()
```

1.Assign the DataFrame

```python
# Oscar number and Nominations columns to the variable 'new'
new = df[['Oscar number', 'Nominations']]

# Language to the variable 'newer'
newer = df.drop(columns=['Language'])

new, newer
```

2. Retrieve the rows of all the movies

```python
# Duration longer than 200 minutes
long_duration_movies = df[df['Duration (minutes)'] > 200]

# Less than 12 Oscar nominations
less_than_12_nominations = df[df['Nominations'] < 12]

long_duration_movies, less_than_12_nominations
df
```

3. Access the 2nd row and the 3rd column.AND Access the columns named Release and Nominations.

```python
# Access the 2nd row and 3rd column
sec_row_thi_clmn = df.iloc[1, 2]

rls_noms = df[['Release', 'Nominations']]

# Displaying the results
sec_row_thi_clmn, rls_noms
```

4. Access the value for the row with index 10 and the column name 'Director' AND Access the value for the row with the index 2 and the column name 'Release'

```python
rowten = df.at[10, 'Director']
rowtwo = df.at[2, 'Release']

rowten, rowtwo
```
